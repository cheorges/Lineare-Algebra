## Skalarprodukt

```python
def dotProduct(X, Y):
    return sum(x * y for x, y in zip(X, Y))
```

```python
dotProduct([1,2],[3,4])
```

## Norm eines Wektors

```python
from math import sqrt

def norm(X):
    return sqrt(dotProduct(X, X))
```

```python
print(f'{norm([0,0])}, {norm([1,1])}, {norm([2,4])}')
```

## Abstand zweier Vektoren

```python
def vectorSub(X, Y):
    return [x - y for x, y in zip(X, Y)]
```

```python
def distance(X, Y):
    return norm(vectorSub(X, Y))
```

```python
distance([4,2], [3,1])
```

## Winkel zweier Vektoren

```python
from math import acos, degrees

def angle(X, Y):
    radians = acos(dotProduct(X, Y)/(norm(X) * norm(Y)))
    return degrees(radians)
```

```python
angle([32,63], [-4,15])
```

## Vektorenprodukt

```python
def crossProduct(X, Y):
    return [X[1] * Y[2] - X[2] * Y[1],
            X[2] * Y[0] - X[0] * Y[2],
            X[0] * Y[1] - X[1] * Y[0]]
```

```python
crossProduct([2,-3,4], [3,-2,-7])
```
