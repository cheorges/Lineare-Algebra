# Skalarprodukt

```python
def dotProduct(X, Y):
    return sum(x * y for x, y in zip(X, Y))
```

```python
dotProduct([1,2],[3,4])
```

## Norm eines Wektors

```python
from math import sqrt

def norm(X):
    return sqrt(dotProduct(X, X))
```

```python
print(f'{norm([0,0])}, {norm([1,1])}, {norm([2,4])}')
```
